```python
# Code from all-MiniLM-L6-v2 hugging face page


from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)


# Sentences we want sentence embeddings for
sentences = ['Elephants are bigs creatures', 'One of the tallest animal is the Elephant']

# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
old_model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')

# Tokenize sentences
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# Compute token embeddings
with torch.no_grad():
    model_output = old_model(**encoded_input)

# Perform pooling
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

# Normalize embeddings
sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def compare(model, dico, name, add_word=False, print_all=False):
    similarities = []
    for key, values in dico.items():
        if type(values) == str:
            input_ = [key, values]
        else:
            input_ = [key] + values
        if add_word:
            input_ = [f"This is a word : {word}" for word in input_]
        # tokenize the input
        encoded_input = tokenizer(input_, padding=True, truncation=True, return_tensors='pt')

        # Compute token embeddings
        with torch.no_grad():
            model_output = model(**encoded_input)
        # Average token embeddings
        embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
        # Normalize embeddings
        embeddings = F.normalize(embeddings, p=2, dim=1)

        similarity = cosine_similarity(embeddings[0, :], embeddings[1:, :])
        mean = similarity.mean().item()

        if print_all:
            plt.figure(figsize=(12, 6))
            plt.bar(values, similarity.numpy(), label="cosine similarities")
            plt.axhline(mean, color='red', label=f"mean = {np.round(mean, 3)}")
            plt.xticks(rotation=60, ha='right')
            plt.tight_layout()
            plt.title(f"{key} {name}")
            plt.legend()
            plt.show()
        
        # cosine similarity between key and values
        similarities.append(mean)

    plt.figure(figsize=(12, 6))
    plt.bar(dico.keys(), similarities, label="cosine similarities")
    mean = np.mean(similarities)
    plt.axhline(mean, color='red', label=f"mean = {np.round(mean, 3)}")
    plt.xticks(rotation=60, ha='right')
    plt.tight_layout()
    plt.title(name)
    plt.legend()
    plt.show()



compare(old_model,warbreaker_data["positive_pairs"], print_all=False, name="Positive Pairs")
```

```python
from sklearn.decomposition import PCA
def monitor_collapse_pca(model, val_loader, device, epoch, num_samples=2000):
    """
    Récupère 'num_samples' phrases, fait l'ACP et affiche le graphique.
    """
    model.eval()
    all_embeddings = []
    count = 0
    
    with torch.no_grad():
        for batch in val_loader:
            sentences = batch[0]
            
            emb = model(sentences, device=device)
            all_embeddings.append(emb.cpu().numpy())
            
            count += len(sentences)
            if count >= num_samples:
                break
    
    embeddings = np.concatenate(all_embeddings, axis=0)
    
    embeddings = embeddings[:num_samples]


    
    n_components = min(50, embeddings.shape[1], embeddings.shape[0])
    pca = PCA(n_components=n_components) 
    pca.fit(embeddings)
    var = pca.explained_variance_ratio_
    
    print(f"[Epoch {epoch+1}] PCA Variance Top 3: {var[0]}, {var[1]}, {var[2]}")
    if var[0] > 0.4:
        print(f"⚠️ ALERTE: Possible Dimensional Collapse (PC1 explique {var[0]*100:.1f}% de la variance)")

    return var[0]

class Model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
        self.model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')

    def forward(self, sentences, device="cuda"):
        # Get sentences embeddings (code from hugging face)
        # Tokenize sentences
        encoded_input = self.tokenizer(sentences, padding=True, truncation=True, return_tensors='pt').to(device)
        # Compute token embeddings
        model_output = self.model(**encoded_input)
        # Perform pooling
        sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
        # Normalize embeddings
        sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)

        return sentence_embeddings

model = Model()
model(["eeeee"], "cpu").shape
```

```python
from torch.optim import AdamW
from torch.optim.lr_scheduler import OneCycleLR
import torch.nn as nn
from tqdm import tqdm
criterion = nn.BCEWithLogitsLoss()
criterion = nn.CrossEntropyLoss()
#criterion = tanh_cross_entropy
min_lr = 1e-6
max_lr = 1e-5
epochs = 5

def train_loop(model, min_lr, max_lr, epochs):
    optimizer = AdamW(
        model.parameters(),
        lr=min_lr,
        weight_decay=1e-3
    )
    print(next(iter(model.parameters())))
    model.train()
    scheduler = OneCycleLR(
        optimizer,
        max_lr=max_lr,                    
        steps_per_epoch=len(train_loader),
        epochs=epochs,               
    )
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    val_loss = []
    std_history = []  
    var_list = []
    train_loss = []
    temperature = 0.01

    for epoch in tqdm(range(epochs)):
        model.train()
        running_loss = 0.0
        running_std = 0.0
        for batch in train_loader:
            
            first = batch[0]
            second = batch[1]
    
            optimizer.zero_grad()

            first_embed = model(first)
            second_embed = model(second)
            
            cosine = torch.flatten(first_embed @ second_embed.T)
            #logits= torch.atanh(cosine)
            logits = torch.abs(cosine) / temperature
            
            labels = torch.flatten(torch.eye(len(batch[0]))).to(device)
            loss = criterion(logits, labels)
            loss.backward()
    
            optimizer.step()
            scheduler.step()      
    
            running_loss += loss.item()
    
        avg_train_loss = running_loss / len(train_loader)
    
        model.eval()
        
        running_loss = 0.0
        
        with torch.no_grad():
            for batch in val_loader:
                first = batch[0]
                second = batch[1]
                
                optimizer.zero_grad()
    
                first_embed = model(first)
                second_embed = model(second)
                
                batch_std = torch.std(first_embed, dim=0).mean()
                running_std += batch_std.item()
                
                cosine = torch.flatten(first_embed @ second_embed.T)
                logits = torch.abs(cosine) / temperature
            
                labels = torch.flatten(torch.eye(len(batch[0]))).to(device)
                
                loss = criterion(logits, labels)

                running_loss += loss.item()
    
        avg_val_loss = running_loss / len(val_loader)
        avg_std = running_std / len(train_loader)  
        
        std_history.append(avg_std)
        print(
            f"Epoch {epoch+1} | "
            f"Train Loss: {avg_train_loss:.4f} | "
            f"Val Loss: {avg_val_loss:.4f} | "
        )
        train_loss.append(avg_train_loss)
        val_loss.append(avg_val_loss)

        var= monitor_collapse_pca(model, val_loader, device, epoch)
        var_list.append(var)
    return train_loss, val_loss,std_history,var_list
```

```python
## Contrastive loss 

import torch.nn.functional as F

def contrastive_loss(z_i, z_j, temperature=0.05):
    """
    z_i, z_j: tensors (batch_size, embedding_dim), L2-normalized
    Computes NT-Xent / InfoNCE loss
    """
    batch_size = z_i.size(0)

    # Normalize just in case
    z_i = F.normalize(z_i, dim=1)
    z_j = F.normalize(z_j, dim=1)

    # Similarity matrix
    sim_matrix = z_i @ z_j.T / temperature  # shape (batch_size, batch_size)
    
    # Positive similarities are on the diagonal
    positives = torch.diag(sim_matrix)

    # Denominator = sum of exp(sim) over all pairs (excluding self if needed)
    exp_sim = torch.exp(sim_matrix)
    denom = exp_sim.sum(dim=1)

    # InfoNCE loss
    loss = - torch.log(torch.exp(positives) / denom)
    return loss.mean()
```

```python
def collapse_score(embeddings):
    # embeddings: (batch, dim)
    std_per_dim = embeddings.std(dim=0)
    mean_std = std_per_dim.mean().item()
    return mean_std

```
